This year you are responsible for helping stage a reenactment of this historic event. The current mayor, Mr. Wimblethumb, has asked you and your pal Jeff to use your spud gun to launch a potato from the center of town square into the public pool as a celebration of the heroic deed. 

You and Jeff mount the spud gun in town square at a 45 degree angle. Getting the potato to land in the pool is a bit hit or miss. Nonetheless, you figure out roughly how much propellant you need to do the job.

You suspect Jeff used too much propellant, but Jeff blames your potato cannon for being unreliable. Can you provide the mayor evidence that Jeff messed up?

## Projectile Model 

To help you figure out if Jeff messed up, you decide to use a simple model of projectile motion with drag. The equations for motion are given by

\begin{equation}
\pmb{F} = m \pmb{g} - c | \pmb{v} |^2 \pmb{v}
\end{equation}
where $\pmb{F}$ is the net force on the spud, $m$ is its mass, $\pmb{v}$ is velocity, and $c$ is a drag coefficient.

```python
import numpy as np
import numba

@numba.jit
def run_sim(v_mag, t_end):
    """ 
    Simulate a projectile with quadratic drag. Runs until the projectile hits the ground.
    """
    # State vector with position and velocity [x, y, v_x, v_y] 
    X = np.zeros(4)
    X[2:] = v_mag*np.array([np.sqrt(2.)/2., np.sqrt(2.)/2.])

    xs = [0.]
    ys = [0.]
    
    # Run the model until the projectile hits the ground
    t = 0
    while t<t_end:
        X += rhs(X)
        t += dt

        xs.append(X[0])
        ys.append(X[1])
        
        if X[1] <= 0.0:
          break

    # Return trajectory
    return xs, ys

# Time step
dt = 0.01
# Gravitational constant
g = 9.81
# Projectile mass
m = 0.1
# Drag coefficient 
c = 1e-5

# RHS of ODE
@numba.jit 
def rhs(X0):

    # Position
    pos = X0[[0,1]]
    # Velocity
    v = X0[[2,3]]
    # Force 
    F = np.array([0., -m*g]) - c*np.linalg.norm(v + 1e-15)**2*v
    
    return np.block([v*dt, (F/m)*dt])

xs, ys = run_sim(200., 20.)

import matplotlib.pyplot as plt

plt.title("The Perfect Shot")
plt.plot(xs, ys)
plt.plot([275., 325.], [0., 0.], 'b-', label = 'Pool')
plt.plot([0.], [0.], 'ko', label = 'Spud Gun')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend()

plt.show()
```
