```python
from itertools import combinations_with_replacement
import numpy as np

n = 4
X = np.zeros((n, 3**n))

indexes = np.arange(n)

r = 0
for i in range(1, n+1):
    index_combs = combinations(indexes, i)
    for index_comb in list(index_combs):
        #print(index_comb)
        sign_combs = combinations_with_replacement(np.array([1.,-1.]), i)
        for sign_comb in list(sign_combs):  
            #print(sign_comb)
            #print(X[index_comb, r])
            X[index_comb, r] = np.sqrt(3.)
            X[index_comb, r] *= sign_comb
            r += 1

        #print()

print(X)
#for c in list(comb1):
#    print(c)

```
